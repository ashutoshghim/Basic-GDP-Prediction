```python
# First we suppress all the warnings that can possibly arise

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing all the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing our data

df = pd.read_csv('data/South_Asian_dataset.csv')
df
```

```python
# Displaying the first few rows of the data
df.head()
```

```python
# We then begin the data cleaning process

df.isnull().sum()
```

```python
# Convert columns with numeric data stored as strings to float
columns_to_convert = [
    'Poverty headcount ratio at $2.15 a day (2017 PPP) (% of population)',
    'Life expectancy at birth, total (years)',
    'Mortality rate, infant (per 1,000 live births)',
    'Literacy rate, adult total (% of people ages 15 and above)',
    'School enrollment, primary (% gross)',
    'Access to electricity (% of population)',
    'People using at least basic drinking water services (% of population)',
    'People using at least basic sanitation services (% of population)',
    'Carbon dioxide (CO2) emissions excluding LULUCF per capita (t CO2e/capita)',
    'PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)',
    'Renewable energy consumption (% of total final energy consumption)',
    'Forest area (% of land area)',
    'Control of Corruption: Percentile Rank',
    'Political Stability and Absence of Violence/Terrorism: Estimate',
    'Regulatory Quality: Estimate',
    'Rule of Law: Estimate',
    'Voice and Accountability: Estimate',
    'Individuals using the Internet (% of population)',
    'Research and development expenditure (% of GDP)',
    'High-technology exports (% of manufactured exports)'
]

for column in columns_to_convert:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Check the data types after conversion
df.dtypes
```

```python
# Let's begin our exploratory data analysis

# Plot GDP growth over the years for each country

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='GDP growth (annual %)', hue='Country')
plt.title('GDP Growth (Annual %) Over the Years')
plt.ylabel('GDP Growth (Annual %)')
plt.xlabel('Year')
plt.legend(title='Country')
plt.show()
```

```python
# Let's calculate the GDP of Nepal

nepal_data = df[df['Country'] == 'Nepal']
nepal_data
```

```python
# Plotting Nepal's growth

plt.figure(figsize=(12, 6))
sns.lineplot(data=nepal_data, x='Year', y='GDP growth (annual %)', hue='Country')
plt.title('GDP Growth (Annual %) Over the Years')
plt.ylabel('GDP Growth (Annual %)')
plt.xlabel('Year')
plt.legend(title='Country')
plt.show()
```

```python
# Plotting the unemployment rates all across the country

plt.figure(figsize=(12, 6))
sns.lineplot(data=nepal_data, x='Year', y='Unemployment, total (% of total labor force) (modeled ILO estimate)', hue='Country')
plt.title('Unemployment Rate Over the Years')
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Year')
plt.legend(title='Country')
plt.show()
```

```python
# Let's conduct a correlation analysis

# Select only numeric columns for correlation analysis
numeric_nepal_data_df = nepal_data.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr = numeric_nepal_data_df.corr()

# Plot the heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Now let's create a simple predictive mechanism after our EDA

# Importing necessary libraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Define the features and target variable
features = numeric_nepal_data_df.drop(columns=['GDP growth (annual %)'])
target = numeric_nepal_data_df['GDP growth (annual %)']
```

```python
# Handle missing values by filling them with the mean of the column
features.fillna(features.mean(), inplace=True)
target.fillna(target.mean(), inplace=True)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```
